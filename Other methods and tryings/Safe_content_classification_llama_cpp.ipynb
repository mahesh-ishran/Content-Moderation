```python
from llama_cpp import Llama
import json, re
```

```python
import os

num_cores = os.cpu_count()
print(f"This machine has {num_cores} CPU cores.")
```

```python
# Path to your downloaded .gguf file
MODEL_PATH = "bartowski/Meta-Llama-3.1-8B-Instruct-GGUF"

llm = Llama(model_path=MODEL_PATH)
```

```python
llm = Llama.from_pretrained(
	repo_id="MaziyarPanahi/Meta-Llama-3.1-8B-Instruct-GGUF",
	filename="Meta-Llama-3.1-8B-Instruct.Q2_K.gguf",
)
```

```python
llm.create_chat_completion(
	messages = [
		{
			"role": "user",
			"content": "What is the capital of France?"
		}
	]
)
```

```python
prompt = """
You are a careful, literal content-safety classifier.
Classify the INPUT text into:
- Explicit Nudity Detection
- Suggestive Content Detection
- Violence Detection
- Rude Gestures Detection
- Alcohol Content Detection
- Drugs Content Detection
- Tobacco Content Detection
- Hate Symbol Detection

Output strictly as JSON with:
{
  "Label": {"flag": 0 or 1, "rationale": "..."}
}

INPUT:
"The dealer was caught selling pills and white powder on the street corner."
"""
```

```python
output = llm(prompt, max_tokens=512, temperature=0)
```

```python
output
```

```python
text_out = output["choices"][0]["text"]
```

```python
# Extract JSON if possible
match = re.search(r"\{.*\}", text_out, re.S)
if match:
    result = json.loads(match.group(0))
    print(json.dumps(result, indent=2))
else:
    print("No JSON output:", text_out)
```
